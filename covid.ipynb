```python
import requests
from bs4 import BeautifulSoup

# URL de la página de la que vamos a hacer scraping
url = "https://datosmacro.expansion.com/otros/coronavirus"

# Realizamos la solicitud HTTP
response = requests.get(url)

# Verificamos si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Parseamos el contenido HTML de la página usando 'html.parser' (sin lxml)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Buscamos todas las tablas en la página
    tables = soup.find_all('table')

    # Imprimimos la cantidad de tablas encontradas
    print(f"Se han encontrado {len(tables)} tablas en la página.")

    # Usamos pandas para leer la primera tabla
    # pd.read_html lee las tablas HTML directamente en un DataFrame
    df = pd.read_html(str(tables[0]), header=0)[0]
    #df.to_csv('covid19.csv',index=False)
else:
    print(f"Error al hacer la solicitud HTTP: {response.status_code}")

```

```python
covid = pd.read_csv('covid19.csv')
covid = covid.rename(columns={'Países': 'pais', 'Fecha': 'fecha_reporte'})
covid['pais'] = covid['pais'].str.strip()
covid['pais'] = covid['pais'].replace(" ","")
covid['pais'] = covid['pais'].str.extract('([a-zA-ZáéíóúÁÉÍÓÚüÜñÑ\s]+)', expand=False)

# Limpiar los nombres de las columnas
df.columns = df.columns.str.strip()
```

```python
# Asignar la región usando el diccionario
covid['Region'] = covid['pais'].str.strip().map(regiones)
covid['Region'] = covid['Region'].astype(str)
```

```python
# Filtrar las filas donde 'Region' es nula
regiones_nulas = covid[covid['Region'].isnull()]

# Ver las filas con valores nulos en la columna 'Region'
print(regiones_nulas['pais'].str.strip())
```

```python
covid = covid.drop(['Incremento Muertos','Incremento Confirmados'], axis=1)
```

```python
covid['Confirmados'] = covid['Confirmados'].str.replace(',', '', regex=False)

# Paso 2: Si hay comas decimales (por ejemplo, 2.518,58), reemplazarlas por punto (2.518.58)
covid['Confirmados'] = covid['Confirmados'].str.replace('.', '', regex=False)  # Eliminar puntos de miles
covid['Confirmados'] = covid['Confirmados'].str.replace(',', '.', regex=False)  # Reemplazar coma decimal por punto

covid['Muertos'] = covid['Muertos'].str.replace(',', '', regex=False)

# Paso 2: Si hay comas decimales (por ejemplo, 2.518,58), reemplazarlas por punto (2.518.58)
covid['Muertos'] = covid['Muertos'].str.replace('.', '', regex=False)  # Eliminar puntos de miles
covid['Muertos'] = covid['Muertos'].str.replace(',', '.', regex=False)  # Reemplazar coma decimal por punto
```

```python
covid.dtypes
```

```python
covid['Confirmados'] = pd.to_numeric(covid['Confirmados'])
covid['Muertos'] = pd.to_numeric(covid['Muertos'])
covid['Recuperados'] = covid['Confirmados'] - covid['Muertos']
```

```python
covid['Confirmados'] = pd.to_numeric(covid['Confirmados'])
# Ahora puedes calcular el Top 10 de confirmados
top_10_confirmados = covid.sort_values(by='Confirmados', ascending=False).head(10)

# Mostrar el top 10 de países con los mayores confirmados
print(top_10_confirmados[['pais', 'Confirmados']])
```

```python
# Agrupar por región y calcular totales
region_summary =  covid.groupby("Region")[["Confirmados"]].sum()

# Crear la gráfica
region_summary.plot(kind="bar", figsize=(10, 6), color=["blue"])

# Personalizar la gráfica
plt.title("Totales de casos confirmados en el mundo", fontsize=16)
plt.ylabel("Total", fontsize=12)
plt.xlabel("Región", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Categorías", fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar la gráfica
plt.tight_layout()
plt.ylim(1.0)
plt.show()
```

```python
covid['Region'] = covid['Region'].astype(str)
covid['Curados'] = pd.to_numeric(covid['Muertos'])
# Calcular el total de recuperaciones por región
muertos_por_region = covid.groupby('Region')['Muertos'].sum().reset_index()

# Calcular el total de muertos en todo el dataset
total_muertos = covid['Muertos'].sum()
# Calcular el porcentaje de muertos por región
muertos_por_region['Porcentaje_Recuperaciones'] = (muertos_por_region['Muertos'] / total_muertos) * 100

# Crear el gráfico de pastel sin los porcentajes visibles
plt.figure(figsize=(10, 10))
# Crear el gráfico de pastel sin los porcentajes visibles dentro
plt.pie(muertos_por_region['Porcentaje_Recuperaciones'], 
        labels=muertos_por_region['Region'],  # Solo mostrar las regiones como etiquetas
        startangle=90, 
        labeldistance=1.2)  # Mover las etiquetas fuera del círculo
  # Añadir la leyenda con los porcentajes
plt.legend(
    muertos_por_region['Region'] + ' - ' + muertos_por_region['Porcentaje_Recuperaciones'].round(1).astype(str) + '%',
    loc='upper left',  # Ajustamos la ubicación de la leyenda
    bbox_to_anchor=(1, 1)  # Mueve la leyenda fuera del gráfico (en el lado derecho)
)
plt.title('Porcentaje de Casos de muertes por Región')
plt.show()
```

```python
# Filtrar datos para Europa (supongo que 'Region' tiene valores como 'Europa' y 'America')
europa = covid[covid['Region'] == 'Europa']

# Ordenar por 'Confirmados' y luego por 'Porcentaje_Recuperados'
df_sorted = europa.sort_values(by='Confirmados', ascending=False)
df = df_sorted.sort_values(by='Porcentaje_Curados', ascending=False)

# Seleccionar los primeros 10 países
top_10 = df.head(10)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10['pais'], top_10['Porcentaje_Curados'], color='green')

# Añadir etiquetas con los valores de la columna 'Porcentaje_Curados'
for bar, value in zip(bars, top_10['Porcentaje_Curados']):
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Posición X
        bar.get_height() + 0.5,            # Posición Y ligeramente por encima de la barra
        f'{value:.1f}%',                    # Formato de la etiqueta
        ha='center',                       # Alinear el texto horizontalmente
        va='bottom'                        # Alinear el texto verticalmente
    )

# Configuración del gráfico
plt.xlabel('País')
plt.ylabel('Porcentaje de Recuperados')
plt.xticks(rotation=45)
plt.title('Top 10 de países más afectados por el COVID en Europa')
plt.show()
```
